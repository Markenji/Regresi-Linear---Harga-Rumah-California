## Linear Regression, Ridge And


```python
#house Pricing Dataset
from sklearn.datasets import fetch_california_housing

```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
housing = fetch_california_housing()
```

```python
type(housing)
```

```python
housing
```

```python
dataset=pd.DataFrame(housing.data)
dataset.columns=housing.feature_names
print(dataset.head())
```

```python
dataset['Price'] = housing.target
```

```python
dataset
```

```python
X = dataset.iloc[:,:-1] ##indepent features
y = dataset.iloc[:,-1] #depedent features
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_predict
lin_reg = LinearRegression()
# remove the scoring parameter
mse = cross_val_predict(lin_reg, X, y, cv=5)
mean_mse = np.mean(mse)
print(mean_mse)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
import numpy as np

lin_reg = LinearRegression()

# Compute negative MSE
mse_scores = cross_val_score(lin_reg, X, y, scoring='neg_mean_squared_error', cv=5)

# Convert negative MSE to positive and compute the mean
mean_mse = np.mean(-mse_scores)

print(mean_mse)
```
